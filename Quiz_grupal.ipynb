```python
import json
import requests
from pprint import pprint
```

```python
fichero = requests.get("https://fp.josedomingo.org/lmgs/u05/ej2.json")
pendientes = json.loads(fichero.text)

e=0
for i in pendientes:
    e+=1
print('hay 'f'{e}' ' pruebas de idiomas descritas en el documento')
```

2.

```python
lst=[]
for i in pendientes:
  if i['Horas']>2:
      lst.append(i['Titulo'])
print('los Títulos de las pruebas de nivel que van a durar más de dos horas son:')
for j in lst:
  print('     ',j)
```

3.

```python
print('URL de clases no presenciales: ')
for i in pendientes:
  if i['TipoFormacion']!='Presencial':
    print(i['URL'])
```

```python
ingreso=input('Ingrese ID: ')
print('Título: ')
for i in pendientes:
  if i['$id']==ingreso:
    print('  ',i['Titulo'])
    dicc=i['Profesorado']
print('Profesores: ')
for j in dicc:
  print('   ',j['NombreCompleto'])
```

```python
dicc= {}
dicc2={}
dicc2['inform'] =[]

for i in pendientes:
  Titulo= i['Titulo']
  dicc2['inform'].append(Titulo)
  dicc['Titulo'] = Titulo
  for j in i['Profesorado']:
    dicc['Profesorado'] = j['NombreCompleto']
    dicc2['inform'].append(j['NombreCompleto'])
    
with open('dicc2.json','w')as file:
  json.dump(dicc2, file, indent=4)
print(dicc2)
```
